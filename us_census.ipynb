```python
import pandas as pd
```

```python
df_pop = pd.read_csv('us_pop2010.csv')
```

```python
df_pop = df_pop[['NAME', 'POP', 'POPDEN']].sort_values('POP', ascending=False)
df_pop = df_pop.set_index(pd.RangeIndex(start=1, stop=3602, step=1))
```

```python
large = df_pop[df_pop['POP'] > 1000000]
medium = df_pop[(df_pop['POP'] <= 1000000) & (df_pop['POP'] > 100000)]
small = df_pop[(df_pop['POP'] <= 100000) & (df_pop['POP'] > 10000)]
micro = df_pop[df_pop['POP'] <= 10000]
```
